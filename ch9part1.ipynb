## <p style="background-color: grey; color: white "> ◼️  Dictionary </p>
# <p style="font-family: Arial; font-size:1.4em;color:red;"> 0. Quick Start — Create, Print, Traverse</p>


````python
# Example: Create a dictionary (a function mapping)
f = {1: 'a', 2: 'b', 3: 'c'}
print(f)
print(type(f))
print(f[1])    # access value by key
print(f[3])    # access value by key
````

A dictionary maps **keys** to **values**. Think of it as a function:
- Keys = inputs (domain)
- Values = outputs (range)
- `f[1]` means "what does f map 1 to?" → `'a'`


````python
# Example: Get keys, values, items
f = {1: 'a', 2: 'b', 3: 'c'}

print('keys:  ', f.keys())      # all inputs
print('values:', f.values())     # all outputs
print('items: ', f.items())      # all (key, value) pairs
````

### Traversing a Dictionary
You can loop through a dictionary by **keys**, by **values**, or by **both**:


````python
# Example: Traverse by key-value pairs
f = {1: 'a', 2: 'b', 3: 'c'}

# Method 1: by keys
print('--- by keys ---')
for k in f:
    print(k, '→', f[k])

# Method 2: by items (key, value)
print('--- by items ---')
for k, v in f.items():
    print(k, '→', v)

# Method 3: by values only
print('--- values only ---')
for v in f.values():
    print(v)
````

### Domain and Range from a Dictionary
In discrete math, for a function `f: X → Y`:
- **Domain** = `set(f.keys())` — all inputs
- **Range** = `set(f.values())` — all outputs actually produced

Using `set()` removes duplicates and gives you a proper mathematical set.


````python
# Example: Domain and Range
f = {1: 'a', 2: 'b', 3: 'c'}
g = {1: 'a', 2: 'b', 3: 'a'}   # note: 1 and 3 both map to 'a'

print('f domain:', set(f.keys()))     # {1, 2, 3}
print('f range: ', set(f.values()))    # {'a', 'b', 'c'}

print('g domain:', set(g.keys()))     # {1, 2, 3}
print('g range: ', set(g.values()))    # {'a', 'b'}  ← only 2 unique outputs
````

### Checking Membership
- `key in f` — checks if a **key** exists
- `val in f.values()` — checks if a **value** exists


````python
# Example: Check membership
f = {1: 'a', 2: 'b', 3: 'c'}

print(1 in f)            # True  — 1 is a key
print(5 in f)            # False — 5 is not a key
print('a' in f.values()) # True  — 'a' is a value
print('z' in f.values()) # False — 'z' is not a value
````

A `dict` (short for dictionary) is a collection of key-value pairs, where each key is unique and is used to access its associated value. It is an unordered, mutable, and indexed collection that allows for fast data retrieval.

### Key Characteristics:
- **Mutable**: You can change the values of existing keys or add new key-value pairs.
- **Unique Keys**: Keys must be unique, though values can be duplicated.
- **Unordered**: Python dictionaries do not maintain the order of items (before Python 3.7), but starting from Python 3.7, they maintain insertion order.

<h1> Accessing Elements in the Dictionary </h1>
<a href="https://docs.python.org/3/library/stdtypes.html#mapping-types-dict"> Python Official Documentation </a>




Accessing elements in a dictionary is done using the keys. You can retrieve the value associated with a specific key using the following syntax:

```python
value = dictionary[key]
```


````python
courses= dict(CRN='CSV17', CourseTitle='Discrete Structures', CourseCredit=3, CourseGrade='A')
courses['CourseTitle']
````

### How to print all element with key value pairs
```python
for key, value in courses.items():
    print(f"{key}: {value}")
```


### How to print all keys
```python
for key in courses.keys():
    print(key)
```


### How to print all values
```python
for value in courses.values():
    print(value)
```


### How to print all values according to the keys
```python
for key in courses.keys():
    print(courses[key])
```

### How to check if a key exists in the dictionary
```python
if 'CourseTitle' in courses:
    print("Yes, 'CourseTitle' is one of the keys in the course_info dictionary")
```

### How to find the key from the value  
```python
list(courses.keys())[list(courses.values()).index('Discrete Structures')]
```
